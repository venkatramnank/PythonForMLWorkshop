## 3.3 Basic Operations in openCV

### 3.3.1 Opening images 

```python
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
img_bgr = cv2.imread('husky.jpg')
plt.imshow(img_bgr)
```

```python
import cv2
img = cv2.imread("husky.jpg")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

### 3.3.2 Resizing Images

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img =cv2.resize(img_rgb,(1300,275))
plt.imshow(img)
```

### 3.3.3 Image Rotation

```python
mg_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# Along central x axis
new_img = cv2.flip(img_rgb,0)
plt.imshow(new_img)

```

## 3.4 Application of OpenCV in Face Detection

```python
img = cv2.imread("photo.jpg")
```

```python
plt.imshow(img)
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
```

```python
gray = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2GRAY)
```

```python
faces = face_cascade.detectMultiScale(gray, 1.1, 4)
```

```python
for (x, y, w, h) in faces:
    cv2.rectangle(img_rgb, (x, y), (x+w, y+h), (255, 0, 0), 2)
```

```python
plt.imshow(img_rgb)
```
