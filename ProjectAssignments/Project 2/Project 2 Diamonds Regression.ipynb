```python
# Understanding the data
data.info()
```

```python
#The first column seems to be just index, thus we remove that column
data = data.drop(["Unnamed: 0"], axis=1)
data.describe()
```

#### Min value of "x", "y", "z" are zero for some of the rows. This indicates that there are missing values in data. So we need to filter out those as it clearly faulty data points.

```python
#Dropping data that have no dimensions
data = data.drop(data[data["x"]==0].index)
data = data.drop(data[data["y"]==0].index)
data = data.drop(data[data["z"]==0].index)
data.shape
```

```python
# check for missing values 
data.isnull().any()
```

```python
ax=sns.pairplot(data, hue= "cut")
```

We see a lot of outliers in the 'x','y','depth','table' and 'z'.

```python
#Dropping the outliers. 
data = data[(data["depth"]<75)&(data["depth"]>45)]
data = data[(data["table"]<80)&(data["table"]>40)]
data = data[(data["x"]<30)]
data = data[(data["y"]<30)]
data = data[(data["z"]<30)&(data["z"]>2)]
data.shape
```

```python
ax=sns.pairplot(data, hue= "cut")
```

```python
# Get list of categorical variables
s = (data.dtypes =="object")
object_cols = list(s[s].index)
print("Categorical variables:")
print(object_cols)
```

```python
# Using label encoding to convert string to numerical values
# Make copy to avoid changing original data 
label_data = data.copy()

# Apply label encoder to each column with categorical data
label_encoder = LabelEncoder()
for col in object_cols:
    label_data[col] = label_encoder.fit_transform(label_data[col])
label_data.head()
```

```python
# Features and Targets : X and y
X= label_data.drop(["price"],axis =1)
y= label_data["price"]
```

```python
# Dividing into train and test data
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.25, random_state=7)
```

```python
# Using Random forest regressor
regr = RandomForestRegressor(max_depth=10, random_state=0)
```

```python
regr.fit(X, y)
```

```python
pred = regr.predict(X_test)
```

```python
# Model Evaluation
print("R^2:",metrics.r2_score(y_test, pred))
print("Adjusted R^2:",1 - (1-metrics.r2_score(y_test, pred))*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1))
print("MAE:",metrics.mean_absolute_error(y_test, pred))
print("MSE:",metrics.mean_squared_error(y_test, pred))
print("RMSE:",np.sqrt(metrics.mean_squared_error(y_test, pred)))
```
